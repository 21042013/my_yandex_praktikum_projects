```python
data_level_finished = data_level_finished.merge(last_data, on = 'user_id', how='inner')
```

Интерес представляет срез данных о пользователях, прошедших за исследуемый период первый уровень.  Структура данных позволяет определить какую стратеги они выбрали (мирное развитие – peace или агрессию - war) исследовать особенности поведения присущие реализации каждой из стратегий.

Визуализация распределений времен прохождения уровней и расчет средних и медиан показывает, что пользователи выбравшие стратегию peace в большей части проходя уровень за более длительный срок ( 13 против 10-11 дней).

Более того, пользователи выбравшие стратегию peace, строят не менее 9 зданий каждый. Результат – более равномерное рекламное воздействие.  Больше людей увидят больше рекламы. 

В среднем пользователе выбравшие стратегию peace строят большее количество ( смотрят больше страниц)  всех типов зданий. На 16% больше сборочных цехов,  на 25% больше космопортов, на 50% больше исследовательских центров в среднем на одного пользователя.

Ежедневная активность по группам (метрика DAU) примерно равна. Из графиков видно, что те пользователи, которые не прошли уровень до конца периода в подавляющем большинстве прекратили игру.  Для уровня 1 retention rate можно грубо оценить как соотношение пользователей прошедших уровень к общему количеству пользователей, что составляет ~ 43%. 

Исходя из вышеизложенного, можно прийти к выводу, что пользователи выбирающие стратегию peace более перспективны для целей монетизации – они строят больше зданий ( просматривают больше целевых страниц), за большее время проходят уровень ( больше возможностей еще как-то образом показать рекламу).

Вот только действующая рекламная стратегия привлекает в основном  пользователей выбирающих стратегию war.
